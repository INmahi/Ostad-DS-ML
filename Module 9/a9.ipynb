```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Quick look
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,10))

# Row 1, Column 1: Math
plt.subplot(2,2,1)
sns.histplot(df['math score'], kde=True, color='skyblue', bins=20)
plt.title("Math Score Distribution")

# Row 1, Column 2: Reading
plt.subplot(2,2,2)
sns.histplot(df['reading score'], kde=True, color='salmon', bins=20)
plt.title("Reading Score Distribution")

# Row 2, Span both columns: Writing
plt.subplot(2,1,2)  # 2 rows, 1 column, position 2 (full width)
sns.histplot(df['writing score'], kde=True, color='green', bins=20)
plt.title("Writing Score Distribution")

plt.tight_layout()
plt.show()
```

```python
fig = px.histogram(df, x=['math score','reading score','writing score'], barmode='overlay',
                   nbins=20, title="Distribution of Scores", labels={'value':'Score','variable':'Subject'})
fig.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='test preparation course', hue='gender', data=df, palette='Set2')
plt.title("Test Preparation Course by Gender")
plt.show()
```

```python
fig = px.histogram(df, x='test preparation course', color='gender', barmode='group',
                   title='Test Preparation Course by Gender')
fig.show()
```

```python
fig = go.Figure(data=go.Heatmap(
                   z=df[['math score','reading score','writing score']].corr().values,
                   x=['Math','Reading','Writing'],
                   y=['Math','Reading','Writing'],
                   colorscale='Viridis',
                   zmin=0, zmax=1))
fig.update_layout(title="Correlation Heatmap of Scores")
fig.show()
```
